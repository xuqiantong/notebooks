```python
from phonemizer import phonemize
from phonemizer.separator import Separator
from phonemizer.backend import EspeakBackend
from phonemizer.punctuation import Punctuation
from phonemizer.logger import get_logger
import fasttext as ft

            
def g2p(inputs):
    model = ft.load_model('/checkpoint/michaelauli/asr/cv-corpus-6.1-16khz/lid.187.bin')
    lang, pnmzer, fn = inputs
    lang = lang.split('-')[0]
    lg_label = f"__label__{lang}"
    print('*** running ', fn, '***')
    with open(fn + '.wrd.raw') as f:
        with open(fn + '.phn.raw', 'w') as fout:
            ok = 0
            notok = 0
            for line in f:
                line = line.strip()
                try:
                    lid, prob = model.predict(line, k=10)
                    if not lg_label in lid:
                        ipa = ""
                        notok += 1
                    else:
                        target_idx = lid.index(lg_label)
                        if target_idx == 0 or prob[target_idx] >= 0.2:
                            ipa = pnmzer.phonemize(
                                line, separator=Separator(phone=' ', syllable='', word='_'), strip=True, njobs=1)
                            ok += 1
                        else:
                            ipa = ""
                            notok += 1
                except:
                    ipa = ""
                    notok += 1
                fout.write(ipa.strip().replace('_', ' ') + '\n')
                fout.flush()
    print('*** finish ', fn, '***', ok, '|', notok)

pnmzer = EspeakBackend(
            'tt',
            punctuation_marks=Punctuation.default_marks(),
            preserve_punctuation=False,
            with_stress=False,
            language_switch='remove-flags',
            logger=get_logger())
g2p(('tt', pnmzer, '/checkpoint/michaelauli/asr/cv-corpus-6.1-16khz/tt/test'))
```

```python
print(len(cv2esp_lang_map))
root = '/checkpoint/michaelauli/asr/cv-corpus-6.1-16khz/'

def get_duration(fn):
    duration = 0.
    with open(fn) as f:
        for i, line in enumerate(f):
            if i == 0:
                continue
            nframes = int(line.split('\t')[-1])
            duration += nframes / 3600
    return str(round(duration / 16000, 2))

outlists = []
suffix = ""
for k, v in cv2esp_lang_map.items():
    outlist = [k, v] + lang_family_map[v]
    outlist.append(get_duration('{}{}/train.tsv{}'.format(root, k, suffix)))
    outlist.append(get_duration('{}{}/dev.tsv{}'.format(root, k, suffix)))
    outlist.append(get_duration('{}{}/test.tsv{}'.format(root, k, suffix)))
    with open('{}{}/dict.phn{}.txt'.format(root, k, suffix)) as f:
        lines = f.readlines()
        outlist.append(str(len(lines)))
    outlists.append(outlist)

outlists = sorted(outlists, key=lambda item: item[3])
for outlist in outlists:
#     print(outlist[0])
    print('\t'.join(outlist))
```

```python
# mls

from phonemizer import phonemize
from phonemizer.separator import Separator
from phonemizer.backend import EspeakBackend
from phonemizer.punctuation import Punctuation
from phonemizer.logger import get_logger
import fasttext as ft

            
def g2p(inputs):
    model = ft.load_model('/checkpoint/michaelauli/asr/cv-corpus-6.1-16khz/lid.187.bin')
    lang, pnmzer, fn = inputs
    lang = lang.split('-')[0]
    lg_label = f"__label__{lang}"
    print('*** running ', fn, '***')
    with open(fn + '.wrd.raw') as f:
        with open(fn + '.phn.raw', 'w') as fout:
            ok = 0
            notok = 0
            for line in f:
                line = line.strip()
                try:
                    lid, prob = model.predict(line, k=10)
                    if not lg_label in lid:
                        ipa = ""
                        notok += 1
                    else:
                        target_idx = lid.index(lg_label)
                        if target_idx == 0 or prob[target_idx] >= 0.2:
                            ipa = pnmzer.phonemize(
                                line, separator=Separator(phone=' ', syllable='', word='_'), strip=True, njobs=1)
                            ok += 1
                        else:
                            ipa = ""
                            notok += 1
                except:
                    ipa = ""
                    notok += 1
                fout.write(ipa.strip().replace('_', ' ') + '\n')
                fout.flush()
    print('*** finish ', fn, '***', ok, '|', notok)

for lang in ['de', 'en', 'es', 'fr', 'it', 'nl', 'pl',  'pt']:
    pnmzer = EspeakBackend(
            lang,
            punctuation_marks=Punctuation.default_marks(),
            preserve_punctuation=False,
            with_stress=False,
            language_switch='remove-flags',
            logger=get_logger())
g2p(('tt', pnmzer, f'/private/home/qiantong/w2v/data/mls/full/{lang}/test'))
```

```python
import panphon.distance
dst = panphon.distance.Distance()
            
measurements = {
    0:dst.fast_levenshtein_distance,
    1:dst.dogol_prime_distance,
    2:dst.feature_edit_distance,
    3:dst.hamming_feature_edit_distance,
    4:dst.weighted_feature_edit_distance,
}

def find_nn(probe_phn, phns, measure_type=3):
    min_d = 100
    res = ""
    for phn in phns:
        measurement = measurements[measure_type]
        d = measurement(probe_phn, phn)
        if d > min_d: continue
        res = phn
        min_d = d    
    return res

def find_knn(probe_phn, phns, measure_type=3, threshold=1):
    res = []
    for phn in phns:
        measurement = measurements[measure_type]
        d = measurement(probe_phn, phn)
        if d > threshold: continue
        res.append((phn, d))
    res = sorted(res, key=lambda x : x[1])
    return res

def gen_lexicon(valid_dict, full_dict, lexicon_path, include_train_oov=False, include_target_oov=False):
    valid_phns = set([])
    with open(valid_dict) as f:
        for line in f:
            valid_phns.add(line.split()[0])
#     print(len(valid_phns), valid_phns)   
    
    full_phns = set([])
    with open(full_dict) as f:
        with open(lexicon_path, 'w') as fout:
            for line in f:
                sp = line.split()
                phn = sp[0].strip()
                full_phns.add(phn)
                
                if phn in valid_phns:
                    fout.write(f'{phn}\t{phn}\n')
                elif include_train_oov:
                    valid_phn = find_knn(phn, valid_phns)
                    fout.write(f'{valid_phn}\t{phn}\n')
                
            if include_target_oov:
                for phn in valid_phns - full_phns:
                    valid_phn = find_knn(phn, full_phns)
                    fout.write(f'{phn}\t{valid_phn}\n')
```

```python
all_lang = ["en","de","nl","fr","es","it","pt"]

target_dict_path = "/private/home/qiantong/w2v/data/new_mls/10h/{}/dict.phn.txt"
train_dict_path = "/private/home/qiantong/w2v/data/new_mls/10h/joint-{}/dict.phn.txt"
lexicon_path = "/private/home/qiantong/w2v/data/new_mls/10h/joint-{}/lexicon_candidate0.txt"

for lang in all_lang:
    target_dict = set()
    with open(target_dict_path.format(lang)) as f:
        for line in f:
            target_dict.add(line.split()[0])
    train_dict = set()
    with open(train_dict_path.format(lang)) as f:
        for line in f:
            train_dict.add(line.split()[0])
         
    with open(lexicon_path.format(lang), 'w') as f:
        for p in target_dict:
            valid_phns = find_knn(p, train_dict, 3, 0.05)
            if len(valid_phns) == 0:
                valid_phns = [find_nn(p, train_dict)]
            for valid_phn in valid_phns:
                f.write(f'{p}\t{valid_phn[0]}\n')

```

```python
import panphon.distance
dst = panphon.distance.Distance()
            
measurements = {
    0:dst.fast_levenshtein_distance,
    1:dst.dogol_prime_distance,
    2:dst.feature_edit_distance,
    3:dst.hamming_feature_edit_distance,
    4:dst.weighted_feature_edit_distance,
}

def find_knn(probe_phn, phns, measure_type=3, threshold=0):
    res = []
    for phn in phns:
        measurement = measurements[measure_type]
        d = measurement(probe_phn, phn)
        if d > threshold: continue
        res.append((phn, d))
    res = sorted(res, key=lambda x : x[1])
    return res

def find_knn(probe_phn, phns, measure_type=3, threshold=1):
    res = []
    for phn in phns:
        measurement = measurements[measure_type]
        d = measurement(probe_phn, phn)
        if d > threshold: continue
        res.append((phn, d))
    res = sorted(res, key=lambda x : x[1])
    return res

target_phns = set([])
with open('/checkpoint/michaelauli/aconneau/datasets/BABEL/new/Georgian/conversational/dict.phn.txt') as f:
    for line in f:
        target_phns.add(line.split()[0])
print(target_phns)

with open(f'/private/home/qiantong/w2v/data/zeroshot_26/babel_Georgia_lexicon.lst', 'w') as fout:
    with open('/private/home/qiantong/w2v/data/zeroshot_26/dict.phn.txt') as f:
        train_phns = set([])
        for line in f:
            phn = line.split()[0]
            train_phns.add(phn)
            if phn in target_phns:
                fout.write(f"{phn}\t{phn}\n")
            else:
                valid_phn = find_nn(phn, target_phns)
                print(valid_phn, phn)
                fout.write(f"{valid_phn}\t{phn}\n")
                
        for phn in target_phns - train_phns:
            valid_phn = find_nn(phn, train_phns)
            print(phn, valid_phn)
            fout.write(f"{phn}\t{valid_phn}\n")
            
print('-----')            
with open(f'/private/home/qiantong/w2v/data/zeroshot_26/babel_Georgia_lexicon_map.lst', 'w') as fout:
    for phn in target_phns:
        valid_phns = find_knn(phn, train_phns, 3, 0)
        print(phn, valid_phns)
        for valid_phn in valid_phns:
            fout.write(f"{phn}\t{valid_phn[0]}\n")
```

```python
import editdistance

root = f"/checkpoint/qiantong/w2v/zero_shot/babel_results"

def get_pter(fn):
    refs = []
    hypos = []
    
    with open(fn) as f:
        for line in f:
            if 'HYPO' in line:
                phns = line.split(':')[-1].strip().split()
                hypos.append(phns)
            elif 'REF' in line:
                phns = line.split(':')[-1].strip().split()
                refs.append(phns)
    
    assert len(refs) == len(hypos)
    
    total_len = 0.
    total_dis = 0.
    for i in range(len(refs)):
        ref = ''.join(refs[i])
        hypo = ''.join(hypos[i])
        
        total_len += len(ref)
        total_dis += editdistance.eval(ref, hypo)
    
    return str(round(total_dis * 100 / total_len, 2))

results = {}
all_lang = ['Cantonese', 'Lao', 'Vietnamese', 'Zulu', 'Georgian']
for l in all_lang:
    results[l] = []
    
for category in ['viterbi', 'ngram']:
    for lang in all_lang:
        fn = f'{root}/{category}/babel_{lang}_valid/infer.log'
        try:
            wer = get_pter(fn)
            results[lang].append(wer)
        except:
            results[lang].append("0")
    
cnt = 0
print('\t'.join(all_lang))
for category in ['viterbi', 'ngram']:
    res = []
    for l in all_lang:
        res.append(str(results[l][cnt]))
    print('\t'.join(res))
    cnt += 1
```

```python
## 2.1 generate map-joint-*

root = '/private/home/qiantong/w2v/data/new_mls/10h'
base_data_path = '/datasets01/mls'

import panphon.distance
dst = panphon.distance.Distance()
            
import random
def find_nn(probe_phn, phns):
    min_d = 100
    res = ""
    for phn in phns:
        d = dst.hamming_feature_edit_distance(probe_phn, phn)
        if d > min_d: continue
        res = phn
        min_d = d    
    return res

def find_knn(probe_phn, phns, threshold=0):
    res = []
    for phn in phns:
        d = dst.hamming_feature_edit_distance(probe_phn, phn)
        if d > threshold: continue
        res.append((phn, d))
    res = sorted(res, key=lambda x : x[1])
    
    if len(res) > 0:
        mind = res[0][1]
        res = [i[0] for i in res if i[1] == mind]
    return res

def map_phns(raw_phns, target_dict):
    raw_phns = lines_phn[i].strip().split()
    phns = []
    for p in raw_phns:
        if p in target_dict:
            phns.append(p)
        elif p in train_phn_map:
            n = len(train_phn_map[p])
            if n == 0: return []
            phns.append(train_phn_map[p][random.randrange(n)])
        else:
            map_p = find_knn(p, target_dict.keys(), 0.05)
            n = len(map_p)
            if n == 0: 
                map_p = [find_nn(p, target_dict.keys())]
            train_phn_map[p] = map_p
            n = len(map_p)
            phns.append(map_p[random.randrange(n)])
            # print("I am OK")
    return phns
    
for lang in ['de', 'es', 'fr', 'it', 'nl', 'pl','pt', 'en']:
    print(lang)
    outpath = f'{root}/map-joint-{lang}'
    os.system(f"mkdir -p {outpath}")
    os.system(f"ln -s {root}/{lang}/dict.phn.txt {outpath}/dict.phn.txt")
    os.system(f"ln -s {root}/{lang}/valid.phn {outpath}/valid.phn")
    os.system(f"ln -s {root}/{lang}/valid.tsv {outpath}/valid.tsv")
    os.system(f"ln -s {root}/{lang}/test.phn {outpath}/test.phn")
    os.system(f"ln -s {root}/{lang}/test.tsv {outpath}/test.tsv")
    
    target_dict = {}
    with open(f'{root}/{lang}/dict.phn.txt') as f:
        for line in f:
            sp = line.split()
            target_dict[sp[0]] = int(sp[1])
            
    with open(f'{outpath}/lexicon.lst', 'w') as fout:
        for k, v in target_dict.items():
            fout.write(f'{k}\t{k}\n') 
    
    train_phn_map = {}
    with open(f'{outpath}/train.phn', 'w') as fphn, open(f'{outpath}/train.tsv', 'w') as ftsv:
        ftsv.write(base_data_path + '\n')
        for ll in ['de', 'es', 'fr', 'it', 'nl', 'pl','pt', 'en']:
            if ll == lang:
                continue
            
            with open(f'{root}/{ll}/train.phn') as f:
                lines_phn = f.readlines()
            with open(f'{root}/{ll}/train.tsv') as f:
                lines_tsv = f.readlines()
                
            tsv_base = lines_tsv[0].strip()
            relpath = os.path.relpath(tsv_base, base_data_path)
            
            for i in range(len(lines_phn)):
                raw_phns = lines_phn[i]
                new_phns = map_phns(raw_phns, target_dict)
                if len(new_phns) == 0: continue
                    
                ftsv.write(f'{relpath}/{lines_tsv[i + 1]}')
                fphn.write(' '.join(new_phns) + '\n')
                
    target_phns = set(target_dict.keys())
    train_mapped_phns = set()
    for _, v in train_phn_map.items():
        for p in v:
            train_mapped_phns.add(p)
    print(len(target_phns), len(train_mapped_phns))
```

```python
import panphon.distance
dst = panphon.distance.Distance()
            
measurements = {
    0:dst.fast_levenshtein_distance,
    1:dst.dogol_prime_distance,
    2:dst.feature_edit_distance,
    3:dst.hamming_feature_edit_distance,
    4:dst.weighted_feature_edit_distance,
}

def find_nn(probe_phn, phns, measure_type=3):
    min_d = 100
    res = ""
    for phn in phns:
        measurement = measurements[measure_type]
        d = measurement(probe_phn, phn)
        if d > min_d: continue
        res = phn
        min_d = d    
    return res

def find_knn(probe_phn, phns, measure_type=3, threshold=1):
    res = []
    for phn in phns:
        measurement = measurements[measure_type]
        d = measurement(probe_phn, phn)
        if d > threshold: continue
        res.append((phn, d))
    res = sorted(res, key=lambda x : x[1])
    return res
```

```python
def gen_lexicon(valid_dict, full_dict, lexicon_path):
    valid_phns = set([])
    with open(valid_dict) as f:
        for line in f:
            valid_phns.add(line.split()[0])
    print('lexicon', len(valid_phns))   
    
    full_phns = set([])
    with open(full_dict) as f:
        with open(lexicon_path, 'w') as fout:
            for line in f:
                sp = line.split()
                phn = sp[0].strip()
                full_phns.add(phn)
                
                valid_phn = find_nn(phn, valid_phns)
                fout.write(f'{valid_phn}\t{phn}\n')
                
            for phn in valid_phns - full_phns:
                valid_phn = find_nn(phn, full_phns)
                fout.write(f'{phn}\t{valid_phn}\n')
                
def gen_map_lexicon(valid_dict, full_dict, lexicon_path):
    full_phns = set([])
    with open(full_dict) as f:
        for line in f:
            full_phns.add(line.split()[0])
    print('map_lexicon', len(full_phns))   
    
    with open(valid_dict) as f:
        with open(lexicon_path, 'w') as fout:
            for line in f:
                sp = line.split()
                phn = sp[0].strip()
                
                valid_phns = find_knn(phn, full_phns, 3, 0)
                for valid_phn in valid_phns:
                    fout.write(f'{phn}\t{valid_phn[0]}\n')
```

```python
## word lexicons espeak
from phonemizer import phonemize
from phonemizer.separator import Separator
from phonemizer.backend import EspeakBackend
from phonemizer.punctuation import Punctuation
from phonemizer.logger import get_logger

def containsNumber(value):
    numbers = re.findall(r'[0-9]+', value)
    return True if numbers else False 

def get_espeak_word_lexicon(lang, phn_dict, all_words, use_map=False):
    pnmzer = EspeakBackend(
            lang,
            punctuation_marks=Punctuation.default_marks(),
            preserve_punctuation=False,
            with_stress=False,
            language_switch='remove-flags',
            logger=get_logger())
    
    with open(f'{outdir}/lexicon_wrd_{lang}.lst', 'w') as f:
        for word in all_words:
            ipa = pnmzer.phonemize(word, separator=Separator(phone=' ', syllable='', word='_'), strip=True, njobs=1)
            valid_ipa = [[]]
            try:
                for phn in ipa.split():
                    if not use_map:
                        if phn not in phn_dict:
                            valid_ipa[0].append(find_nn(phn, phn_dict))
                        else:
                            valid_ipa[0].append(phn)
                    else:
                        tmp_valid_ipa = []

                        valid_phns_raw = find_knn(phn, phn_dict, 3, 1)
                        valid_phns = []
                        for p in valid_phns_raw:
                            if not containsNumber(p[0]):
                                valid_phns.append(p)
                        k = 0
                        while valid_phns[k][1] == valid_phns[0][1]: k += 1
                        k = min(k, 3)
                        for tmp in valid_ipa:
                            for p in valid_phns[:k]:
                                tmp_valid_ipa.append(tmp + [p[0]])
                        valid_ipa = copy.deepcopy(tmp_valid_ipa)
            except:
                continue

#             print(word, len(valid_ipa), valid_ipa)
            for ipa in valid_ipa:
                if len(ipa) == 0: continue
                spelling = ' '.join(ipa)
                f.write(f'{word}\t{spelling}\n')
                f.flush()
            
#             break
    

outdir = '/private/home/qiantong/w2v/data/espeak_25lang_m10'
phn_dict = set()
with open(f'{outdir}/dict.phn.txt') as f:
    for line in f:
        phn_dict.add(line.split()[0])

# for lang in ['tt', 'ky']:
for lang in ['it']:
    all_words = set()
    for part in ['train', 'dev']:
        with open(f'/checkpoint/michaelauli/asr/cv-corpus-6.1-16khz/{lang}/{part}.wrd') as f:
            for line in f:
                words = line.strip().split()
                for w in words:
                    all_words.add(w)
                    
    get_espeak_word_lexicon(lang, phn_dict, all_words, False)
#     get_espeak_word_lexicon(lang, phn_dict, all_words, True)
```

```python
## word lexicons phonetisaurus
import re

def remove_stress(phn):
    phn = phn.replace('ʼ', '')
    phn = phn.replace('ˈ', '')
    phn = phn.replace('.', '')
    phn = phn.replace('*', '')
    phn = phn.replace('~', '')
    phn = phn.replace('˞', '')
    phn = phn.replace('#', '')
    phn = re.sub('˥', ' ', phn)
    phn = re.sub('˩', ' ', phn)
    phn = re.sub('˧', ' ', phn)
    phn = re.sub('˨', ' ', phn)
    phn = re.sub('˦', ' ', phn)
    phn = re.sub(' +', ' ', phn)
    return phn.strip()
            
def get_phonetisaurus_word_lexicon(lang, phn_dict, all_words, use_map=False):
    with open(f'/tmp/{lang}_words.lst', 'w') as fout:
        for w in all_words:
            fout.write(w + '\n')
    
    cmd = f"""
    /private/home/qiantong/w2vu/local/bin/phonetisaurus-g2pfst \
    --model={fst_model_base}/{fst_models[lang]} \
    --print_scores=false --wordlist=/tmp/{lang}_words.lst > /tmp/{lang}_phns.lst
    """
    os.system(cmd)
    
    lexicon = {}
    with open(f'/tmp/{lang}_phns.lst') as f:
        for line in f:
            sp = line.strip().split('\t')
            try:
                phns = remove_stress(sp[1])
            except:
                continue
            if len(phns) == 0:
                continue

            valid_ipa = []
            for phn in phns.strip().split():
                if not use_map:
                    if phn not in phn_dict:
                        valid_ipa.append(find_nn(phn, phn_dict))
                    else:
                        valid_ipa.append(phn)
                else:
                    pass
            lexicon[sp[0]] = ' '.join(valid_ipa)
        
    return lexicon

outdir = '/private/home/qiantong/w2v/data/phonetisaurus_40lang_m10'
phn_dict = set()
with open(f'{outdir}/dict.phn.txt') as f:
    for line in f:
        phn_dict.add(line.split()[0])

for lang in ['tt', 'ky']:
    all_words = set()
    for part in ['train', 'dev']:
        with open(f'/checkpoint/michaelauli/asr/phonetisaurus_no_flag/{lang}/{part}.wrd') as f:
            for line in f:
                words = line.strip().split()
                for w in words:
                    all_words.add(w)
                    
    lexicon = get_phonetisaurus_word_lexicon(lang, phn_dict, all_words)
    with open(f'{outdir}/lexicon_wrd_{lang}.lst', 'w') as f:
        for k, v in lexicon.items():
            if len(v) > 0:
                f.write(f'{k}\t{v}\n')
```

# PER threshold

```python
import panphon.distance
dst = panphon.distance.Distance()
            
measurements = {
    0:dst.fast_levenshtein_distance,
    1:dst.dogol_prime_distance,
    2:dst.feature_edit_distance,
    3:dst.hamming_feature_edit_distance,
    4:dst.weighted_feature_edit_distance,
}

dis_dic = {}
with open('/tmp/viterbi_tt/dev_tt/infer.log') as f:
    refs = []
    hypos = []
    for line in f:
        if 'HYPO' in line:
            phns = line.split(':')[-1].strip().split()
            hypos.append(phns)
        elif 'REF' in line:
            phns = line.split(':')[-1].strip().split()
            refs.append(phns)
            
    phn_set = set([])
    assert len(refs) == len(hypos)
    for i in range(len(refs)):
        for p in refs[i]:
            phn_set.add(p)
        for p in hypos[i]:
            phn_set.add(p)
            
    measurement = measurements[measure_type]
    for p1 in phn_set:
        for p2 in phn_set:
            d = measurement(p1, p2)
            dis_dic[f'{p1}_{p2}'] = d
            
print(dis_dic)
```

```python
measure_type = 3
threshold = 0.

def isSamePhn(p1, p2):
    return dis_dic[f'{p1}_{p2}'] <= threshold


def levenshteinDistance(s1, s2):
    if len(s1) > len(s2):
        s1, s2 = s2, s1

    distances = range(len(s1) + 1)
    for i2, c2 in enumerate(s2):
        distances_ = [i2+1]
        for i1, c1 in enumerate(s1):
            if isSamePhn(c1, c2):
                distances_.append(distances[i1])
            else:
                distances_.append(1 + min((distances[i1], distances[i1 + 1], distances_[-1])))
        distances = distances_
    return distances[-1]

    
total_len = 0.
total_dis = 0.
for i in range(len(refs)):
    ref = refs[i]
    hypo = hypos[i]
        
    total_len += len(ref)
    total_dis += levenshteinDistance(ref, hypo)
    
print(str(round(total_dis * 100 / total_len, 2)))
```
